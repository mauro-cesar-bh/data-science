# Practical Exam: House sales

RealAgents is a real estate company that focuses on selling houses.

RealAgents sells a variety of types of house in one metropolitan area.

Some houses sell slowly and sometimes require lowering the price in order to find a buyer.

In order to stay competitive, RealAgents would like to optimize the listing prices of the houses it is trying to sell.

They want to do this by predicting the sale price of a house given its characteristics.

If they can predict the sale price in advance, they can decrease the time to sale.


## Data

The dataset contains records of previous houses sold in the area.

| Column Name | Criteria                                                |
|-------------|---------------------------------------------------------|
| house_id    | Nominal. </br> Unique identifier for houses. </br>Missing values not possible. |
| city        | Nominal. </br>The city in which the house is located. One of 'Silvertown', 'Riverford', 'Teasdale' and 'Poppleton'. </br>Replace missing values with "Unknown". |
| sale_price  | Discrete. </br>The sale price of the house in whole dollars. Values can be any positive number greater than or equal to zero.</br>Remove missing entries. |
| sale_date   | Discrete. </br>The date of the last sale of the house. </br>Replace missing values with 2023-01-01. |
| months_listed  | Continuous. </br>The number of months the house was listed on the market prior to its last sale, rounded to one decimal place. </br>Replace missing values with mean number of months listed, to one decimal place. |
| bedrooms    | Discrete. </br>The number of bedrooms in the house. Any positive values greater than or equal to zero. </br>Replace missing values with the mean number of bedrooms, rounded to the nearest integer. |
| house_type   | Ordinal. </br>One of "Terraced" (two shared walls), "Semi-detached" (one shared wall), or "Detached" (no shared walls). </br>Replace missing values with the most common house type. |
| area      | Continuous. </br>The area of the house in square meters, rounded to one decimal place. </br>Replace missing values with the mean, to one decimal place. |

```python
house_sales = pd.read_csv('house_sales.csv')

# Check if all house_id values are unique
house_sales_unique = house_sales['house_id'].is_unique
print("House IDs are unique:", house_sales_unique)


# City -> Define names and replace the rest by Unknow
cities = ['Silvertown', 'Riverford', 'Teasdale', 'Poppleton', 'Unknown']
# Ensuring all values are valid categories, replacing invalid ones with 'Unknown'
house_sales['city'] = house_sales['city'].where(house_sales['city'].isin(cities), 'Unknown')
# # Converting the 'region' column to categorical with the predefined categories
house_sales['city'] = pd.Categorical(house_sales['city'], categories=cities)

# Check if there is nulls or numbers below 0 by looking at the stats
house_sales = house_sales[~house_sales['sale_price'].isna()]
print(house_sales['sale_price'].describe())

# Check if there is nulls or numbers below 0 by looking at the stats
# house_sales['sale_date'] = pd.to_datetime(house_sales['sale_date'], errors='coerce').dt.date
house_sales['sale_date'].fillna('2023-01-01', inplace=True)
print(house_sales['sale_date'].describe())

#  Fill null values with mean rounded in one decimal phase
mean_months = house_sales['months_listed'].mean()
house_sales['months_listed'].fillna(round(mean_months, 1), inplace=True)
house_sales['months_listed'] = house_sales['months_listed'].round(1)
print("Mean months listed:", mean_months)

# Bedrooms -> Check if there is nulls or numbers below 0 by looking at the stats
house_sales['bedrooms'] = house_sales['bedrooms'].where(house_sales['bedrooms']>= 0)
mean_bedrooms = house_sales['bedrooms'].mean()
house_sales['bedrooms'].fillna(round(mean_bedrooms, 0), inplace=True)
print(house_sales['bedrooms'].describe())

#  House Type -> 
house_type = ['Detached', 'Semi-detached', 'Terraced']
# Dictionary for exact matches
replacement_dict = {
    'Det.': 'Detached',
    'Terr.': 'Terraced',
    'Semi': 'Semi-detached'
}
house_sales['house_type'] = house_sales['house_type'].replace(replacement_dict)

house_mode = house_sales['house_type'].mode()[0]
print("House type most common:", house_mode)
house_sales['house_type'] = house_sales['house_type'].where(house_sales['house_type'].isin(house_type), house_mode)
house_sales['house_type'] = pd.Categorical(house_sales['house_type'], categories=house_type,ordered=True)

# #  House Area -> 
# Step 1: Remove the suffix ' sq.m'
house_sales['area'] = house_sales['area'].str.replace(' sq.m.', '').astype(float)
# Step 2: Convert the cleaned strings to float
house_sales['area'] = pd.to_numeric(house_sales['area'], errors='coerce').round(1)
mean_area = house_sales['area'].mean()
house_sales['area'].fillna(round(mean_area, 1), inplace=True)

null_house_prices = house_sales['area'].isnull().any()
print("There are null areas:", null_house_prices)

# house_sales['area'] = house_sales['area'].astype(float).round(1)
clean_data = house_sales.copy()
print(clean_data.info())
clean_data
# clean_data[['area_2', 'area']]
```

# Task 3 

The team at RealAgents have told you that they have always believed that the number of bedrooms is the biggest driver of house price. 

Producing a table showing the difference in the average sale price by number of bedrooms along with the variance to investigate this question for the team.

 - You should start with the data in the file 'house_sales.csv'.

 - Your output should be a data frame named `price_by_rooms`. 

 - It should include the three columns `bedrooms`, `avg_price`, `var_price`. 

 - Your answers should be rounded to 1 decimal place.   

```python
house_sales = pd.read_csv('house_sales.csv')
price_by_rooms = house_sales.groupby('bedrooms')['sale_price'].agg(['mean', 'var']).round(1).rename(columns={'mean': 'avg_price', 'var': 'var_price'})

price_by_rooms = price_by_rooms.reset_index()
price_by_rooms
```

```python
import datetime as dt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

train_data = pd.read_csv('train.csv')
validation_data = pd.read_csv('validation.csv')

# Tranform data in ordinal 
train_data['sale_date'] = pd.to_datetime(train_data['sale_date'])
train_data['sale_date_ordinal'] = train_data['sale_date'].map(dt.datetime.toordinal)

validation_data['sale_date'] = pd.to_datetime(validation_data['sale_date'])
validation_data['sale_date_ordinal'] = validation_data['sale_date'].map(dt.datetime.toordinal)


# Assuming 'spend' is the target variable and other columns are features
target = 'sale_price'
X = train_data.drop(columns=[target, 'sale_date'])
y = train_data[target]


# If there are categorical variables, convert them to dummy variables or use encoding
categorical_cols = ['city', 'house_type']
X = pd.get_dummies(X, columns=categorical_cols)
X_test = pd.get_dummies(validation_data, columns=categorical_cols)



# Align test data with training data (in case of missing columns)
X, X_test = X.align(X_test, join='left', axis=1, fill_value=0)

# Initialize the model
model = LinearRegression()

#  Split
# X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# # Fit the model on training data
# model.fit(X_train, y_train)

# # Predict on validation set (optional)
# y_val_pred = model.predict(X_val)

# # Compute RMSE
# rmse = mean_squared_error(y_val, y_val_pred, squared=False)
# print("RMSE: {}".format(rmse))

# Not using cross val -> Fit the model on training data
model.fit(X, y)

#  Predict Valyes
test_price_pred =  model.predict(X_test)

# Create a result DataFrame for predictions
base_result = pd.DataFrame({
    'house_id': validation_data['house_id'],
    'price': test_price_pred
})

base_result
```
